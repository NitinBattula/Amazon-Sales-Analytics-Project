```python
import pandas as pd
```

```python
salesdf = pd.read_csv(r"C:\Users\Nitin\Documents\DA Projects\amazon_sales_data 2025.csv")     # Reading csv file
salesdf   
```

```python
salesdf.rename(columns={'Order ID':'Order_ID','Total Sales':'Total_Sales','Customer Name':'Customer_Name','Customer Location':'Customer_Location','Payment Method':'Payment_Method'},inplace=True)
salesdf                                       # Renaming the necessary columns
```

```python
salesdf.info()
```

```python
salesdf['Date']=pd.to_datetime(salesdf['Date'])       #Changing the data type of Date
salesdf.info()
```

```python
salesdf['Date']=pd.to_datetime(salesdf['Date'],format='%Y-%m-%d')       # Changing the format of Date
salesdf
```

```python
salesdf.insert(1,'Year',salesdf['Date'].dt.year)
salesdf.insert(2,'Month_Name',salesdf['Date'].dt.month_name())
salesdf.insert(3,'Month',salesdf['Date'].dt.month)                               #Adding Year,Month_name,Month,Day columns
salesdf.insert(4,'Day',salesdf['Date'].dt.day)
salesdf
```

```python
salesdf.drop(columns='Date',inplace=True)                      #Dropping the Date column
salesdf
```
